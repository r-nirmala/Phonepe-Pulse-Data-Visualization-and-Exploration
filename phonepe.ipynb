```python
import pandas as pd
import os
import json
import mysql.connector
```

```python
#dataframe of aggregated transaction
path = 'C:/Users/rnirm/OneDrive/Documents/Guvi Data Science - MDE86/Guvi Projects/Project_2/Phonepe_dataset/pulse/data/aggregated/transaction/country/india/state/'
agg_tran  = os.listdir(path)
columns = {'State' : [],'Year' : [],'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],
            'Transaction_amount': []}
for state in agg_tran:
    cur_state =path+state+"/"
    agg_year_list= os.listdir(cur_state)
    for year in agg_year_list:
        cur_year = cur_state+year+"/"
        agg_file_list = os.listdir(cur_year)
        for file in agg_file_list:
            cur_file = cur_year+file
            data = open(cur_file,'r')
            A = json.load(data)
            for i in A['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount =i['paymentInstruments'][0]['amount']
                columns['Transaction_type'].append(name)
                columns['Transaction_count'].append(count)
                columns['Transaction_amount'].append(amount)
                columns['State'].append(state)
                columns['Year'].append(year)
                columns['Quarter'].append(int(file.strip('.json')))
df_agg_tran = pd.DataFrame(columns)
df_agg_tran["State"] = df_agg_tran["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar islands")
df_agg_tran["State"] = df_agg_tran["State"].str.replace("-"," ")
df_agg_tran["State"] = df_agg_tran["State"].str.title()
df_agg_tran['State'] = df_agg_tran['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli & Daman and Diu")
```

```python
#dataframe of aggregated user
path1 = 'C:/Users/rnirm/OneDrive/Documents/Guvi Data Science - MDE86/Guvi Projects/Project_2/Phonepe_dataset/pulse/data/aggregated/user/country/india/state/'
agg_user  = os.listdir(path1)
columns1 = {'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Transaction_Count': [],
            'Percentage': []}
for state in agg_user:
    cur_state =path1+state+"/"
    agg_year_list= os.listdir(cur_state)
    for year in agg_year_list:
        cur_year = cur_state+year+"/"
        agg_file_list = os.listdir(cur_year)
        for file in agg_file_list:
            cur_file = cur_year+file
            data = open(cur_file,'r')
            B = json.load(data)
            try:
                for i in B['data']['usersByDevice']:
                    brand_name = i["brand"]
                    counts = i["count"]
                    percent = i["percentage"]
                    columns1["Brands"].append(brand_name)
                    columns1["Transaction_Count"].append(counts)
                    columns1["Percentage"].append(percent)
                    columns1["State"].append(state)
                    columns1["Year"].append(year)
                    columns1["Quarter"].append(int(file.strip('.json')))
            except:
                pass
            
df_agg_user = pd.DataFrame(columns1)
df_agg_user["State"] = df_agg_user["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar islands")
df_agg_user["State"] = df_agg_user["State"].str.replace("-"," ")
df_agg_user["State"] = df_agg_user["State"].str.title()
df_agg_user['State'] = df_agg_user['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli & Daman and Diu")           
```

```python
#dataframe of map transaction
path2 = 'C:/Users/rnirm/OneDrive/Documents/Guvi Data Science - MDE86/Guvi Projects/Project_2/Phonepe_dataset/pulse/data/map/transaction/hover/country/india/state/'
map_tran = os.listdir(path2)
columns2 = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Transaction_Count': [],
            'Transaction_Amount': []}
for state in map_tran:
    cur_state = path2 + state + '/'
    map_year = os.listdir(cur_state)
    for year in map_year:
        cur_year = cur_state + year + '/'
        map_file = os.listdir(cur_year)
        for file in map_file:
            cur_file = cur_year + file
            data = open(cur_file,'r')
            C = json.load(data)
            for i in C['data']['hoverDataList']:
                district = i['name']
                count = i['metric'][0]['count']
                amount = i['metric'][0]['amount']
                columns2['District'].append(district)
                columns2['Transaction_Count'].append(count)
                columns2['Transaction_Amount'].append(amount)
                columns2['State'].append(state)
                columns2['Year'].append(year)
                columns2['Quarter'].append(int(file.strip('.json')))
df_map_tran = pd.DataFrame(columns2)
df_map_tran["State"] = df_map_tran["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar islands")
df_map_tran["State"] = df_map_tran["State"].str.replace("-"," ")
df_map_tran["State"] = df_map_tran["State"].str.title()
df_map_tran['State'] = df_map_tran['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli & Daman and Diu")           
```

```python
#dataframe of map_user
path3 = 'C:/Users/rnirm/OneDrive/Documents/Guvi Data Science - MDE86/Guvi Projects/Project_2/Phonepe_dataset/pulse/data/map/user/hover/country/india/state/'
map_user = os.listdir(path3)
columns3 = {"State": [], "Year": [], "Quarter": [], "District": [],
            "RegisteredUser": [], "AppOpens": []}
for state in map_user:
    cur_state = path3 + state + "/"
    map_year = os.listdir(cur_state)
    
    for year in map_year:
        cur_year = cur_state + year + "/"
        map_file = os.listdir(cur_year)
        
        for file in map_file:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            D = json.load(data)
            for i in D["data"]["hoverData"].items():
                district = i[0]
                registereduser = i[1]["registeredUsers"]
                appOpens = i[1]['appOpens']
                columns3["District"].append(district)
                columns3["RegisteredUser"].append(registereduser)
                columns3["AppOpens"].append(appOpens)
                columns3['State'].append(state)
                columns3['Year'].append(year)
                columns3['Quarter'].append(int(file.strip('.json')))
df_map_user = pd.DataFrame(columns3)
df_map_user["State"] = df_map_user["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar islands")
df_map_user["State"] = df_map_user["State"].str.replace("-"," ")
df_map_user["State"] = df_map_user["State"].str.title()
df_map_user['State'] = df_map_user['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli & Daman and Diu")           

            
```

```python
#Dataframe of top transaction
path4 = 'C:/Users/rnirm/OneDrive/Documents/Guvi Data Science - MDE86/Guvi Projects/Project_2/Phonepe_dataset/pulse/data/top/transaction/country/india/state/'
top_tran =os.listdir(path4)

columns4 = {'State':[], 'Year':[], 'Quarter':[], 'Pincode':[], 'Transaction_count':[], 'Transaction_amount':[]}

for state in top_tran:
    cur_state = path4+state+"/"
    top_year = os.listdir(cur_state)
    
    for year in top_year:
        cur_year = cur_state+year+"/"
        top_file = os.listdir(cur_year)
        
        for file in top_file:
            cur_file = cur_year+file
            data = open(cur_file,"r")
            E = json.load(data)
            for i in E['data']['pincodes']:
                name = i['entityName']
                count = i['metric']['count']
                amount = i['metric']['amount']
                columns4['Pincode'].append(name)
                columns4['Transaction_count'].append(count)
                columns4['Transaction_amount'].append(amount)
                columns4['State'].append(state)
                columns4['Year'].append(year)
                columns4['Quarter'].append(int(file.strip('.json')))    
df_top_tran = pd.DataFrame(columns4)
df_top_tran["State"] = df_top_tran["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar islands")
df_top_tran["State"] = df_top_tran["State"].str.replace("-"," ")
df_top_tran["State"] = df_top_tran["State"].str.title()
df_top_tran['State'] = df_top_tran['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli & Daman and Diu")           
            
```

```python
#Dataframe of top user
path5 = 'C:/Users/rnirm/OneDrive/Documents/Guvi Data Science - MDE86/Guvi Projects/Project_2/Phonepe_dataset/pulse/data/top/user/country/india/state/'
top_user = os.listdir(path5)

columns5 = {'State':[], 'Year':[], 'Quarter':[], 'Pincode':[],'RegisteredUsers':[]}

for state in top_user:
    cur_state = path5+state+'/'
    top_year = os.listdir(cur_state)
    for year in top_year:
        cur_year = cur_state+year+'/'
        top_file = os.listdir(cur_year)
        for file in top_file:
            cur_file = cur_year+file
            data = open(cur_file,'r')
            F = json.load(data)
            for i in F['data']['pincodes']:
                name = i['name']
                registeredUsers = i['registeredUsers']
                columns5['Pincode'].append(name)
                columns5['RegisteredUsers'].append(registeredUsers)
                columns5['State'].append(state)
                columns5['Year'].append(year)
                columns5['Quarter'].append(int(file.strip('.json')))
df_top_user = pd.DataFrame(columns5)

df_top_user["State"] = df_top_user["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar islands")
df_top_user["State"] = df_top_user["State"].str.replace("-"," ")
df_top_user["State"] = df_top_user["State"].str.title()
df_top_user['State'] = df_top_user['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli & Daman and Diu")           
```

```python
#MySQL connection
connection = mysql.connector.connect(host="localhost",user="root",password="12345",database="Phonepe")
mycursor = connection.cursor()
```

```python
#Aggr_tran table
mycursor.execute('''create table if not exists agg_tran(State varchar(100), Year int, Quarter int,
                 	Transaction_type varchar(100),	Transaction_count bigint, Transaction_amount bigint)''')
connection.commit()

for i,row in df_agg_tran.iterrows():
    query = '''insert into agg_tran(State, Year, Quarter,
                Transaction_type, Transaction_count, Transaction_amount) 
                values(%s,%s,%s,%s,%s,%s)'''
    mycursor.execute(query,tuple(row))
    connection.commit()
```

```python
#Aggr_user table
mycursor.execute('''create table if not exists agg_user(State varchar(100), Year int, Quarter int,
                 Brands varchar(100), Transaction_Count bigint, Percentage float)''')
connection.commit()

for i,row in df_agg_user.iterrows():
    query = '''insert into agg_user(State, Year, Quarter, Brands, Transaction_Count, Percentage) 
                values(%s,%s,%s,%s,%s,%s)'''
    mycursor.execute(query,tuple(row))
    connection.commit()
```

```python
#map_tran table

mycursor.execute('''create table if not exists map_tran(State varchar(100), Year int, Quarter int,
                 District varchar(100), Transaction_Count bigint, Transaction_Amount float)''')
connection.commit()

for i,row in df_map_tran.iterrows():
    query = '''insert into map_tran(State, Year, Quarter, District, Transaction_Count, Transaction_Amount) 
                values(%s,%s,%s,%s,%s,%s)'''
    mycursor.execute(query,tuple(row))
    connection.commit()
```

```python
#map_user table

mycursor.execute('''create table if not exists map_user(State varchar(100), Year int, Quarter int,
                 District varchar(100), RegisteredUser bigint, AppOpens bigint)''')
connection.commit()

for i,row in df_map_user.iterrows():
    query = '''insert into map_user(State, Year, Quarter, District, RegisteredUser, AppOpens) 
                values(%s,%s,%s,%s,%s,%s)'''
    mycursor.execute(query,tuple(row))
    connection.commit()
```

```python
#top_tran table

mycursor.execute('''create table if not exists top_tran(State varchar(100), Year int, Quarter int,
                 Pincode int, Transaction_count bigint, Transaction_amount float)''')
connection.commit()

for i,row in df_top_tran.iterrows():
    query = '''insert into top_tran(State, Year, Quarter, Pincode, Transaction_count, Transaction_amount) 
                values(%s,%s,%s,%s,%s,%s)'''
    mycursor.execute(query,tuple(row))
    connection.commit()
```

```python
#top_user table

mycursor.execute('''create table if not exists top_user(State varchar(100), Year int, Quarter int,
                 Pincode int, RegisteredUsers bigint)''')
connection.commit()

for i,row in df_top_user.iterrows():
    query = '''insert into top_user(State, Year, Quarter, Pincode, RegisteredUsers) 
                values(%s,%s,%s,%s,%s)'''
    mycursor.execute(query,tuple(row))
    connection.commit()
```
